### Whenever you are dealing with categorical data ENCODED as numbers (for example Q1, Q2,Q3,Q4 represented as 1,2,3,4)
### Plotly will make errors while interpreting column's type in coloring ans use a color scale


### Solution : Convert such encoded columns to categorical format! 

```python
# Plot a bar graph to show average closing price for each quarter in 2017

data = df.loc['2017'].resample('Q') [['Close']].mean()

data.index = pd.Categorical ( data.index.quarter ) #manual conversion to avoid plotly express errors


fig =   pe.bar( 
    x=data.index,
    y='Close',
    data_frame=data,
    color=data.index,
    text_auto=True  )

#update the figure
fig.update_layout(

    #xaxis changes are to be made
    xaxis=dict(

        tickmode = 'array', #array means x axis markers are categorical in nature
        tickvals=[1,2,3,4], #tickvals indicates the positions we want to edit on xaxis
        ticktext = ['Q1','Q2','Q3','Q4'] #pass new label(ticks) values
    ),

    yaxis=dict(

        tickmode = 'linear', #y axis is numerical
        tick0 = 0, #lowest value of marker on y axis
        dtick = 10 #distance between 2 consecutive tick markers
    ),


    title='Average Close Price (2017-Quarterly)',

    xaxis_title = 'Quarter',

    yaxis_title = 'Average Close Price',

    height=600
)

#width needs to be between 0 and 1 : Think of this as factor for width
fig.update_traces(width=0.25)

fig.show()
```
