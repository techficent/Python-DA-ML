# Clean Data

```python
# rename columns

df.columns = [     col.title() for col in df.columns         ]


df
```

```python
# drop irrelevant column

df.drop(columns=['Bmi','Id'], inplace=True)

df
```

## Separate categorical & numerical feature

```python
categorical_features = ['Gender', 'Hypertension', 'Heart_Disease','Ever_Married','Work_Type','Residence_Type','Smoking_Status']

numerical_features=['Age','Avg_Glucose_Level']
```

# Check if dataset is biased to any specific label class


```python
df['Stroke'].value_counts(normalize=True)
```

# Encode categorical features

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder() 

for col in categorical_features:
    df[col]   =  le.fit_transform(  df[col]  )


df
```

# Split features & labels

```python
X = df[numerical_features].copy()

y=df[['Stroke']].copy()

from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test = train_test_split( X, y,random_state=10, stratify=y )
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

# Test the model

```python
ans = model.predict(X_test)

predicted_df = pd.DataFrame(ans, columns=['Prediction'])

predicted_df
```

```python
test_features_df = X_test.reset_index(drop=True)

actual_test_labels = y_test.reset_index(drop=True)

pd.concat([test_features_df,actual_test_labels, predicted_df],axis=1)
```

# Metrics


```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,accuracy_score,confusion_matrix

print(precision_score(  y_test, predicted_df))

print(recall_score(  y_test, predicted_df))

print(accuracy_score(y_test,predicted_df))


print(confusion_matrix(y_test,predicted_df))
```
