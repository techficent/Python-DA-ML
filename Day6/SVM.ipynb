# Load data 

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/techficent/Python-DA-ML/master/Day6/healthcare-dataset-stroke-data.csv")

df
```

### Exploration of data

```python
# Check missing value count


df.isna().sum()
```

```python
# Check data type of each column and identify categorical v/s numerical data columns

print(df.info())


print("#####Number of unique values in each column")
for col in df.columns:
    print(f"{col} ----> {df[col].nunique()}  ")
```

# Clean Data

```python
# rename columns

df.columns = [     col.title() for col in df.columns         ]


df
```

# Check if dataset is biased to any specific label class


```python
df['Stroke'].value_counts(normalize=True)
```

# Split features & labels

```python
X = df[['Avg_Glucose_Level', 'Age']].copy()

y=df[['Stroke']].copy()

from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test = train_test_split( X, y,random_state=10, stratify=y, test_size=0.2 )
```

# If you are solving classification problem, keep stratify option enabled!

# Select the algorithm

```python
from sklearn.svm import SVC


#support vector classifier
model = SVC()
```

## Fit the model on the data!

```python
model.fit(X_train,y_train) # training the model!
```

# Supervised learning

# Test the model

```python
ans = model.predict(X_test)

predicted_df = pd.DataFrame(ans, columns=['Prediction'])

predicted_df
```

# Scaling is required for regression techniques only!

```python
test_features_df = X_test.reset_index(drop=True)

actual_test_labels = y_test.reset_index(drop=True)

pd.concat([test_features_df,actual_test_labels, predicted_df],axis=1)
```

## performance of a classification model???


Model says----> 0


Actual answer is -----> 1

```python
from sklearn.metrics import accuracy_score


accuracy_score(y_test, predicted_df)
```

```python
Trial 


Correct Answer            Model

0                            0               ------> TN (true negative)

0                            1              ----->   FP (false positives)

 
1                            0              -------> FN

1                            1                ------> TP
```

# Metrics


```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,accuracy_score,confusion_matrix

print(precision_score(  y_test, predicted_df))

print(recall_score(  y_test, predicted_df))

print(accuracy_score(y_test,predicted_df))


print(confusion_matrix(y_test,predicted_df))
```

```python
print(confusion_matrix(y_test,predicted_df, labels=[1, 0]))

# TP                FP

# FN                TN
```
