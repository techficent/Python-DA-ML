```python
ans = []


for column in df.columns:
    ans.append(column.title())


df.columns = ans
```

```python

#df.columns should be equal to a list where
    #every item in the list is converted title case for column name present already

df.columns = [    column.title() for column in df.columns         ]


#list comprehension

df
```
