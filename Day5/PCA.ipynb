# Features required for training can be  by using PCA

```python
import pandas as pd

link = "https://raw.githubusercontent.com/techficent/Python-DA-ML/main/timeseries_dummy.csv" 

df = pd.read_csv( link,index_col='Date', parse_dates=['Date'] )

df
```

```python
df.fillna(method='ffill', inplace=True)

df
```

```python
feature = df[[      'Open', 'High', 'Low', 'Adj Close'      ]]

label = df[['Close']]
```

```python
from sklearn.preprocessing    import StandardScaler

sc  =  StandardScaler() 
#fit the data available and then transform(scale) as per z scale formula
scaled_feature = sc.fit_transform(feature)

scaled_feature
```

```python
from sklearn.decomposition import PCA

pca_model = PCA(   n_components=1  )

new_feature = pca_model.fit_transform(  scaled_feature  )

new_feature
```

```python
pca_model.explained_variance_ratio_
```
