# Default Arguments

```python
def additionDefaultArg(n1=10, n2=20):
    print(n1 + n2)
```

```python

#first 3 --->positional arguments
additionDefaultArg()        #10 + 20

additionDefaultArg(7)       #7  + 20

additionDefaultArg(5,15)    #5 + 15


#keyword argument
additionDefaultArg(n2=15)    #10 + 15
```
