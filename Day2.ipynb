```python
#create a list of brands
brands = ['Audi','BMW','Volvo']

#first condition for brands
condition1=(df['Manufacturer'].isin(brands) )

#second condition for price range
condition2=(df['Price_in_thousands'].between(10,50)   )

#create intermediate temp_df after applying both conditions
temp_df = df[  condition1 & condition2       ]

#list of columns to be aggregated
columns=['Price_in_thousands','Year_resale_value']

#apply aggregation min,max & sum
temp_df [  columns  ].agg( ['min','max','sum']  )
```

# Average price of vehicles for each manufacturer

###1) USE GROUPBY with a list of columns
###2) Identify relevant columns for analysis
###3) Deciding aggregate operation to be performed

```python
temp = df.groupby(    ['Manufacturer']        )

avg_df = temp [  ['Price_in_thousands']  ].mean()


avg_df.rename(  columns={'Price_in_thousands': 'Average Price'},inplace=True )

avg_df
```

```python
# Find out number of data points in each group (based on manufacturer)

temp.size()
```
